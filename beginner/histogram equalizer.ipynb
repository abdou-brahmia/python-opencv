```python
import matplotlib.pyplot as plt
#import matplot

a=[790,1023,850,656,329,245,122,81]
#declare our histogram H (i) , i belongs to [0,7]

somme=sum(a)
#calculate the sum of all histogram values

b=[]
#declare array

for i in range(len(a)):
    b.append(a[i]/somme)
    # calculate Hn(i) =H(i)/(sum of all histogram values)
c=[]

#calculate C(i) normalized probability density
for i in range(len(b)):
    if i==0:
        c.append(b[i])
    else:
        c.append(b[i]+c[i-1])
        
    
# transformation 
for i in range(len(c)):
    c[i]=int(c[i]*8)
    
print(c)

# histogram H(i)
plt.plot(a)
plt.xlim([0,7])
plt.show()

# histogram equalizer
plt.plot(c)
plt.xlim([0,7])
plt.show()
```
